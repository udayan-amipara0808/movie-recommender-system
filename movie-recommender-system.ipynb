```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('movies.csv')
```

```python
credits = pd.read_csv('credits.csv')
```

```python
poster = pd.read_csv('poster.csv')
```

```python
#Merge the two datasets and remove unnecessary parallel working.
movies = movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
movies.head()
#movie_id
#title
#overview
#genres
#keywords
#cast
#crew
#these are the column which we require and we have to removw other unnecessary columns.
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies = movies.merge(poster,on='title')
```

```python
movies.head()
```

```python
#Now we have to adjust this text columns like : genres , keywords , cast and crew and have to extract information from them
#genres --> extract keywords of the movies
#keywords --> extract the basic words
#cast --> extract top3 lead actors
#crew --> extract only name of director
```

```python
import ast #library which helped me to convert stirng to list.
```

```python
#helper functions-->
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L

#for extracting top3 actors from movie
def convert3(text):
    L=[]
    counter=0
    for i in ast.literal_eval(text):
        if(counter<3):
            L.append(i['name'])
        counter+=1
    return L
#for getting the name of the Director
def get_director(text):
    L = []
    for i in ast.literal_eval(text):
        if(i['job']=='Director'):
            L.append(i['name'])
    return L

        
        
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres']= movies['genres'].apply(convert)
```

```python
movies['keywords']= movies['keywords'].apply(convert)
```

```python
movies['cast']= movies['cast'].apply(convert3)
```

```python
movies['crew']= movies['crew'].apply(get_director)
```

```python
movies.head() # you can see we converted all the text columns into some meaningful information!!!
```

```python
#we need to remove spaces between the name like model can mistake between recommneding (Sam Mendes) and (Sam Worthington).
def remove(L):
    L1=[]
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(remove)
movies['crew'] = movies['crew'].apply(remove)
movies['genres'] = movies['genres'].apply(remove)
movies['keywords'] = movies['keywords'].apply(remove)
```

```python
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
movies.head(1)
```

```python
#Now i merged a 5 columns overview , genres , keywords , cast and crew and formed a new column "tags".
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies_new = movies.drop(columns = ['overview','genres','keywords','cast','crew'])
```

```python
movies_new.head()
```

```python
movies_new['tags'] = movies_new['tags'].apply(lambda x : " ".join(x))
```

```python
movies_new['tags'] = movies_new['tags'].apply(lambda x : x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    L = []
    for i in text.split():
        L.append(ps.stem(i))
    return " ".join(L)
```

```python
movies_new['tags'] = movies_new['tags'].apply(stem)
```

```python
#Extracted top 5000 Most Frequent words from the tags of each movie so that i can apply Text-Vectorization to it and convert it into a Vector.
# And can Recommend it to a User and can find Vectors closest to it to Recommend to user.
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words = 'english') 
#used stop-words parameter to remove common and grammatical words like it,in,the,a,an etc.
```

```python
vec = cv.fit_transform(movies_new['tags']).toarray()
```

```python
#now i calculated the distance between all the movies but not the euclidean distance but the cosine distance.
#because in Higher Dimension the euclidean distance not give much better results.
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(vec)
```

```python
#movie Recommending function:
def recommend(movie_name):
    movie_index = movies_new[movies_new['title']==movie_name].index[0]
    distances = similarity_matrix[movie_index]
    Top_5 = sorted(list(enumerate(distances)),reverse=True,key = lambda x : x[1])[1:6]
    for i in Top_5:
        print(movies_new.iloc[i[0]].title)
```

```python
recommend('Iron Man')
```

```python
import pickle
pickle.dump(movies_new,open('movies.pkl','wb'))
pickle.dump(similarity_matrix,open('similariaty.pkl','wb'))
```
